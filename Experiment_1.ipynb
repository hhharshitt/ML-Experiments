```python
#Experiment-1 Basic Array Creation
```

```python
import numpy as np
my_list = [10, 20, 30, 40, 50]
arr = np.array(my_list)
print("Array:", arr)
print("Shape:", arr.shape)
print("Data Type:", arr.dtype)
```

```python
#Experiment-2 Zeros and Ones Matrices
```

```python
zeros_matrix = np.zeros((3, 4))
ones_matrix = np.ones((2, 5))

print("Zeros Matrix:\n", zeros_matrix)
print("Shape of Zeros:", zeros_matrix.shape)
print("\nOnes Matrix:\n", ones_matrix)
print("Shape of Ones:", ones_matrix.shape)
```

```python
#Expeiment-3 Reshaping Arrays
```

```python
original = np.arange(1, 13)
reshaped = original.reshape(3, 4)

print("Original Array:", original)
print("Original Shape:", original.shape)
print("\nReshaped Matrix:\n", reshaped)
print("Reshaped Shape:", reshaped.shape)
```

```python
#Experiment 4: Random Integers and Statistics
```

```python
rand_arr = np.random.randint(1, 101, 10)

print("Random Array:", rand_arr)
print("Min:", rand_arr.min())
print("Max:", rand_arr.max())
print("Mean:", rand_arr.mean())
print("Sum:", rand_arr.sum())
```

```python
#Experiment 5: DataFrame from Dictionary
```

```python
import pandas as pd
data = {
    'Student Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Marks': [85, 92, 78, 88]
}

df = pd.DataFrame(data)

print("DataFrame:\n", df)
print("\nColumn Data Types:\n", df.dtypes)
```
