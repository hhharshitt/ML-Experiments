```python
#Experiment 1: Identifying Missing Values
```

```python
csv_content = """StudentID,Name,Math_Score,Science_Score,Attendance
101,Arjun,85,78,95
102,Deepa,92,85,98
103,Sanjay,,80,85
104,Aditi,88,92,90
105,Vijay,76,70,80
106,Manoj,95,,88
107,Neha,89,88,92
108,Rohan,,75,87
109,Priya,82,81,94
110,Amit,90,84,91"""

with open('data.csv', 'w') as f:
    f.write(csv_content)


import pandas as pd
df = pd.read_csv('data.csv')

print("Null values:\n", df.isnull())
print("\nNot null values:\n", df.notnull())
print("\nMissing values count per column:\n", df.isnull().sum())
```

```python
#Experiment 2: Dropping Missing Values
```

```python
print(f"Original Shape: {df.shape}")
df_cleaned = df.dropna()

print(f"Cleaned Shape: {df_cleaned.shape}")
```

```python
#Experiment 3: Filling Missing Values (Imputation)
```

```python
import numpy as np
df['Grade'] = ['A', 'B', np.nan, 'A', 'C', np.nan, 'B', 'A', np.nan, 'B']
df['Grade'] = df['Grade'].fillna("Not Assigned")

print(df[['Name', 'Grade']].head())
```

```python
#Experiment 4: Type Conversion
```

```python
print("Before:\n", df.dtypes)
df['Attendance'] = df['Attendance'].astype(float)

print("\nAfter:\n", df.dtypes)
```

```python
#Experiment 5 & 6: Renaming and Replacing
```

```python
df = df.rename(columns={'Math_Score': 'Mathematics', 'Science_Score': 'Science'})
df['Name'] = df['Name'].replace('Arjun', 'Arjun Kumar')

print(df[['Name', 'Mathematics']].head())
```

```python
#Experiment 7: Line and Bar Plots
```

```python
import matplotlib.pyplot as plt
plt.plot(df['StudentID'], df['Mathematics'], marker='o')
plt.title("Trend of Math Scores")
plt.show()

df.plot(kind='bar', x='Name', y='Science')
plt.title("Science Scores Comparison")
plt.show()
```

```python
#Experiment 8: Scatter Plot (Relationships)
```

```python
plt.scatter(df['Mathematics'], df['Science'])
plt.xlabel("Math Score")
plt.ylabel("Science Score")
plt.title("Relationship between Math and Science")
plt.show()
```

```python
#Experiment 9: Histogram (Distribution)
```

```python
plt.hist(df['Attendance'], bins=5, color='skyblue', edgecolor='black')
plt.title("Attendance Distribution")
plt.show()
```

```python
#Experiment 10: Boxplot (Outliers)
```

```python
plt.boxplot(df['Mathematics'].dropna())
plt.title("Boxplot of Math Scores")
plt.show()
```
